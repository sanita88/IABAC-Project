# ENCODING

```python
cat_cols.columns
```

```python
# Dropping irrelevant colums
data.drop(['EmpNumber'],axis=1,inplace=True)
```

```python
data['Gender'].value_counts()
```

```python
data['Gender']=data['Gender'].replace({'Female':0,'Male':1})
```

```python
data['EducationBackground'].value_counts()
```

```python
data['EducationBackground']= data['Education'] = data['EducationBackground'].map({'Life Sciences': 5,'Medical': 4, 'Marketing': 3,'Technical Degree':2,'Other':1,'Human Resources':0})
```

```python
data['MaritalStatus'].value_counts()
```

```python
data['MaritalStatus']=data['MaritalStatus'].replace({'Married':2,'Single':1,'Divorced':0})
```

```python
data['EmpDepartment'].value_counts()
```

```python
data['EmpDepartment']=  data['EmpDepartment'].map({'Sales': 5,'Development': 4, 'Research & Development': 3,'Human Resources':2,'Finance':1,'Data Science':0})
```

```python
data['EmpJobRole'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['EmpJobRole']=lb.fit_transform(data['EmpJobRole'])
```

```python
data['BusinessTravelFrequency'].value_counts()
```

```python
data['BusinessTravelFrequency']=data['BusinessTravelFrequency'].map({'Travel_Rarely': 2,'Travel_Frequently': 1,'Non-Travel': 0})
```

```python
data['OverTime'].value_counts()
```

```python
data['OverTime']=data['OverTime'].map({'No':1,'Yes':0})
```

```python
data['Attrition'].value_counts()
```

```python
data['Attrition']=data['Attrition'].map({'No':1,'Yes':0})
```
