# Top 3 Important Factors effecting employee performance

**Feature Selection by correlation**

```python
plt.figure(figsize = [20, 15])
sns.heatmap(data.corr(),annot=True)
```

```python
# Calculate the correlation between all columns and 'PerformanceRating'
correlation_matrix = data.corr()
performance_correlation = correlation_matrix['PerformanceRating'].abs()
performance_correlation
```

```python
# Exclude 'PerformanceRating' itself from the list
performance_correlation = performance_correlation.drop('PerformanceRating')

# Select the top 3 factors with the highest absolute correlation
top_3_factors = performance_correlation.nlargest(3)

print(top_3_factors)
```

**Chi-Square Test for Feature Selection**

```python
data1=pd.read_excel('preprocessed_data.xls')
```

```python
x = data1.drop(['PerformanceRating'],axis = 1)
y = data1['PerformanceRating']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

bestfeatures = SelectKBest(score_func=chi2, k=3)
fit = bestfeatures.fit(x,y)
```

```python
scores = pd.DataFrame(fit.scores_)
columns = pd.DataFrame(x.columns)
     
```

```python
scores_fea = pd.concat([columns,scores],axis=1)
scores_fea.columns = ['Specifications','Score']
```

```python
scores_fea.sort_values(by='Score',ascending=False)
```

 SELECTING 3 FEATURES HAVIN HIGH SCORES

```python
print(scores_fea.nlargest(3,'Score'))
```
